# Rocket Trace Data Analysis
This notebook aims to analyze log data recorded by previous ARIS launches and drop tests.

Note: trace data-files are huge and thus not stored in the repository. Please download them from the ARIS sharepoint or Teams and adjust the TRACE_PATH variable in the next cell.

## Analysis of trace data
The trace data is stored in a JSONL file. It includes all data logged during the flight.

```python
import pandas as pd
from pathlib import Path

BASE_PATH = Path("../../external/ARIS-flight-logs")
FILENAME = "trace.jsonl"
TRACE_PATH = BASE_PATH / Path("Aris_Cargoflight_1_Logs") / FILENAME

chunks = pd.read_json(path_or_buf=TRACE_PATH, lines=True, chunksize=100_000)

trace_df = next(chunks)
trace_df["timestamp"] = pd.to_datetime(trace_df["timestamp"])
trace_df = trace_df.sort_values("timestamp").reset_index(drop=True)
print(f"loaded {len(trace_df):,} lines")
```

```python
trace_df.head(n=20)
```

```python
fields_df = pd.json_normalize(trace_df['fields'])
unique_message_titles = fields_df["message"].unique().tolist()

print(f"There are {len(unique_message_titles)} unique message titles")

for i, msg in enumerate(unique_message_titles):
    print(f"{i:3}. {msg}")
```

```python
sensor_df = fields_df[fields_df["message"] == "sending MAVLink message"]
sensor_df.head(20)
```
