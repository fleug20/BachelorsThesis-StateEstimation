# 02 ASTERIA Evaluation

This notebook can be used to evaluate data generated by the 02_asteria_simulation (ballistic descent).

```python
import plt

from core import SimulationResult
from evaluation import Evaluator
from evaluation.aligners import NearestTimeIndexAligner
from evaluation.metrics import PositionRMSE
from evaluation.metrics.attitude_rmse import AttitudeRMSE
from evaluation.metrics.velocity_rmse import VelocityRMSE
from evaluation.runners.rust_runner_asteria import RustRunnerAsteria
from evaluation.visualizer import EvaluationVisualizer

SIMULATION_NAME: str = "02_asteria_ballistic"
DATA_PATH = "../../simulated_data/rocketpy/" + SIMULATION_NAME
```

## Data Loading


```python
simulation_result = SimulationResult.load(DATA_PATH + "/" + SIMULATION_NAME + ".pkl")
```

## Evaluation

```python
from evaluation.metrics import NEES

runner = RustRunnerAsteria(
    use_noisy=True,
    do_correction=True,
    adaptive_gnss=None,
)

aligner = NearestTimeIndexAligner()
metrics = [PositionRMSE(), VelocityRMSE(), AttitudeRMSE(), NEES()]
evaluator = Evaluator(runner, metrics, aligner)
```

```python
result = evaluator.evaluate(simulation_result)
```

```python
result.summary()
```

## Visualization

```python
visualizer = EvaluationVisualizer(result)
visualizer.plot_state_errors()
plt.show()
```

## Export

```python
result.save("../../evaluation_data/rocketpy/" + SIMULATION_NAME + ".pkl")
```
